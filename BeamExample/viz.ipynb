```python
import apache_beam as beam
from apache_beam.runners.interactive.interactive_runner import InteractiveRunner
import apache_beam.runners.interactive.interactive_beam as ib

# Define dummy DoFn classes for visualization
class ParsePubSubMessage(beam.DoFn):
    def process(self, element):
        yield (element, 1)

class FormatOutput(beam.DoFn):
    def process(self, element):
        yield str(element)

# Create pipeline
p = beam.Pipeline(InteractiveRunner())

# Single pipeline chain - using Create instead of ReadFromText for visualization
result = (
        p
        | 'Create Data' >> beam.Create([])
        | 'Parse Messages' >> beam.ParDo(ParsePubSubMessage())
        | 'Apply Window' >> beam.WindowInto(beam.window.FixedWindows(10))
        | 'Group and Count' >> beam.CombinePerKey(sum)
        | 'Format Output' >> beam.ParDo(FormatOutput())
        | 'Print Results' >> beam.Map(print)
)

# Visualize the pipeline graph (does not execute the pipeline)
ib.show_graph(p)
```
